### 1 Attendance Data

Load the attendance.csv file and calculate an attendnace percentage for each student. One half day is worth 50% of a full day, and 10 tardies is equal to one absence.

You should end up with something like this:


- name
- Billy       0.5250
- Jane        0.6875
- John        0.9125
- Sally       0.7625
- Name: grade, dtype: float64

```python
att_df = pd.read_csv('untidy-data/attendance.csv')
att_df.head()
```

```python
att_df.shape
```

```python
att_df.info()
```

```python
att_df.columns
```

```python
att_long = att_df.melt(id_vars=['Unnamed: 0'], var_name='Day', value_name='Score')
att_long.head()
```

```python
att_long.columns = ['name', 'date', 'attendance' ]
att_long.head()
```

```python
# P = 1
# H = .5
# T = .90
# A = 0


score_att = {
    'P': 1,
    'H': 0.5,
    'T': 0.9,
    'A': 0,
}
```

```python
att_long['score_att'] = att_long.attendance.map(score_att)
```

```python
att_long.head()
```

```python
att_long.drop(columns= 'attendance', inplace= True)
```

```python
att_long.groupby('name').score_att.mean()
```

```python
new_df =att_long.groupby('name').score_att.mean()
new_df
```

### 3) Cake Recipes

- a) Read the cake_recipes.csv data. This data set contains cake tastiness scores for combinations of different recipes, oven rack positions, and oven temperatures.
- b) Tidy the data as necessary.
- c) Which recipe, on average, is the best? recipe b
- d) Which oven temperature, on average, produces the best results? 275
- e) Which combination of recipe, rack position, and temperature gives the best result? recipe b, bottom rack, 300 degrees

```python
cake_recipes = pd.read_csv('untidy-data/cake_recipes.csv')
cake_recipes
```

```python
cake_recipes.shape
```

```python
cake_long = cake_recipes.melt(id_vars=['recipe:position'], var_name = 'temperature')
cake_long
```

```python
cake_long[['recipe','position']] = cake_long['recipe:position'].str.split(':', expand = True)
```

```python
cake_long.head()
```

```python
cake_long.drop(columns= 'recipe:position', inplace= True)
```

```python
cake_long.head()
```

```python
cake_pivot = cake_long.pivot_table(index = ['recipe', 'position'],
                       columns = 'temperature',
                       values = 'value')
cake_pivot
```

```python
cake_pivot.columns.name=""
cake_pivot.reset_index()
```

```python
cake_by_rec = cake_pivot.groupby('recipe').mean()
```

```python
cake_by_rec
```

```python
cake_by_rec.T
```

```python
cake_by_rec.T.mean()
```
